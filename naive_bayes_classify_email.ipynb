```python
import re

def mailprocess(text):
    text= text.replace('\r',' ')
    text = text.replace('\n', ' ')
    text = text.replace('#', '')
    text = text.replace("Subject:","")
    text = text.replace("we 're", "we are")
    text = text.replace("ect","etc")
    text = text.replace("you 're","you are")
    return text
    
```

```python
mail_db['clean_text']=mail_db['text'].map(mailprocess)
mail_db
```

## converting clean_text column to vector

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(ngram_range=(1,1))
count= count_vect.fit_transform(mail_db['clean_text'])
print(count.shape)
```

# Dividing data into train/test splits

```python
from sklearn.model_selection import train_test_split
```

```python
x=count
y=mail_db['label_num'].values
print(y)
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
```

# Using Naive Bayes 

```python
gNB = GaussianNB()
```

```python
y_pred=gNB.fit(x_train.toarray(), y_train).predict(x_test.toarray())
print(classification_report(y_test, y_pred))
```

# Confusion matrix

```python
cm=confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sns
pyplt.figure(figsize=(5,5))
sns.heatmap(data=cm,linewidths=1.0,fmt="d", annot=True, square = True, cmap='Blues')
pyplt.ylabel('Actual label')
pyplt.xlabel('Predicted label')

all_sample_title = 'Accuracy Score: {0:.2%}'.format(round(gNB.score(x_test.toarray(), y_test),2))
pyplt.title(all_sample_title, size = 15)
pyplt.savefig("confusion_matrix.png")
```
